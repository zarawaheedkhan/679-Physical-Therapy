```python
# Read data

FPCA = pd.read_csv(wf('Tables/FPCA_mse_r2.csv'))
MFPCA = pd.read_csv(wf('Tables/MFPCA_mse_r2.csv'))
AE = pd.read_csv(wf('Tables/AE_mse_r2.csv')).apply(pd.to_numeric)
VAE = pd.read_csv(wf('Tables/VAE_mse_r2.csv')).apply(pd.to_numeric)

FPCA = FPCA.apply(pd.to_numeric)
MFPCA = MFPCA.apply(pd.to_numeric)
AE = round(AE.apply(pd.to_numeric), 2)
VAE = VAE.apply(pd.to_numeric)
```
